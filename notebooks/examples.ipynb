# 可視化例

## 折れ線グラフ

```python
# matplotlib
x = np.array([0,1,2,3])
y = np.array([3,7,4,8])

plt.plot(x,y)
plt.show()
```

```python
# plotly
data = [go.Scatter(x=x, y=y, yaxis="y2")]
fig = go.Figure(data=data)
offline.iplot(fig)
```

## 散布図

```python
# matplotlib
x = np.array([0,1,2,3])
y = np.array([3,7,4,8])

plt.scatter(x,y)
plt.show()
```

```python
# plotly
trace = go.Scatter(
    x = x,
    y = y,
    mode = 'markers'
)

data = [trace]

# Plot and embed in ipython notebook!
offline.iplot(data, filename='basic-scatter')
```

## 曲線のグラフ

```python
# matplotlib
x = np.linspace(-5, 5, 300)
y = x**2

plt.plot(x,y)
plt.show()
```

```python
# plotly
data = [go.Scatter(x=x, y=y, yaxis="y2")]
fig = go.Figure(data=data)
offline.iplot(fig)
```

## 複数の線を表示する

```python
# matplotlib
x = np.linspace(-5, 5, 300)
y1 = x**2
y2 = (x-2)**2

plt.plot(x,y1)
plt.plot(x,y2,linestyle="--")
plt.show()
```

```python
# plotly
# Create a trace
trace0 = go.Scatter(
    x = x,
    y = y1
)
trace1 = go.Scatter(
    x = x,
    y = y2,
    line = dict(dash = 'dash')
)

data = [trace0, trace1]
fig = go.Figure(data=data)
offline.iplot(fig)
```

## ヒストグラム

```python
np.random.seed(0)
l = []
for _ in range(1000):
    l.append(np.random.randint(1,7,size=10).sum())

plt.hist(l, 20, color="gray")
plt.show()
```

```python
data = [go.Histogram(x=l,histnorm='probability')]
offline.iplot(data)
```
